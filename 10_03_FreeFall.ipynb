# Mass point movement in gravitational field with air resistance



## Free wall with air resistance

Movement equation:

$m\ddot{x}=mg - \frac{1}{2}C_x\rho \dot{x}^2S$

where:

$C_x$ drag coefficient,

$\rho$ density of the fluid(air),

$S$ cross section area.

### Integrate the equation:

$\ddot{x}=g - \frac{1}{2}C_x\rho \dot{x}^2S \frac{1}{m}$



```python
import numpy as np
import pylab as plt

g   = 9.81 # m/s2
Cx  = 0.3  # -
rho = 1.2  # kg/m3
S   = 1    # m2
m   = 80   # kg
#==============================================================================
# simulation
dt = 0.1
steps = int(20/dt) # number of steps for given time in seconds

q = 0        # x
q_dot = 0    # v
q_2dot = 0   # a
t, a, v, x, f = [], [], [], [], []
for i in range(steps):
    """
    Runge-Kutta first order -> Euler
    """
    q_2dot = g - 0.5 * q_dot**2 * Cx * rho * S / m
    q_dot = q_dot + q_2dot * dt
    q = q + q_dot * dt
    
    a.append(q_2dot)
    v.append(q_dot)
    x.append(q)
    t.append(dt*i)
    
#==============================================================================
# plot results
plt.plot(t, v)
plt.title(r'$\ddot{x}=g - \frac{1}{2}C_x\rho \dot{x}^2S \frac{1}{m}$')
plt.xlabel(u'Time [s]')
plt.ylabel(r'Velocity $\left[\frac{m}{s}\right]$')
plt.figure()

plt.plot(t, x)
# 
# nice syntactic sugar
plt.title(r'$\ddot{x}=g - \frac{1}{2}C_x\rho \dot{x}^2S \frac{1}{m}$')
plt.xlabel(u'Time [s]')
plt.ylabel(u'Distance [m]')
plt.legend(['Euler'])
# 
# show
plt.show()
```
