# ODE

Example of usage an ODE solver on simplest motion equation of mass-point moving without friction or environment resistance with single force applied and no gravitational force.

The method __solve_ivp__ from __Scipy__ module is used see (https://docs.scipy.org/doc/scipy/reference/generated/scipy.integrate.solve_ivp.html).

$F=ma$

$m$ - Mass of the mass point [kg]

$a$ - acceleration [$ms^{-2}$]

$F$ - Force applied [N]

$t$ - time of simmulation [s]

```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

# constants
m = 1.0
F = 100.0

initial_vals = [0, 0] # position, velocity

def mass_point(t, Y) :
    x_dot = Y[1]    # velocity
    v_dot = F / m   # acceleration
    return [x_dot, v_dot]

# Solve function mass_point whithin time range 0 to 10 seconds with default
# integration method Runge-Kutta 5(4) with maximum time step limited to 0.1 sec,
# default value is inf, so its very rough integration even in this simple example
# see manual https://docs.scipy.org/doc/scipy/reference/generated/scipy.integrate.solve_ivp.html
data = solve_ivp(mass_point, [0, 10], initial_vals, method='RK45', max_step = 0.1)

# plot results based on time
plt.plot(data.t, data.y[0])
plt.plot(data.t, data.y[1])
plt.xlabel('Time [s]')
plt.legend(['Position [m]', 'Velocity [$ms^{-1}$]'])
plt.title('Mass point movement in time $F=ma$')
plt.show()
```
